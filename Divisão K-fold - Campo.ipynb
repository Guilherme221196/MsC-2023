```python
#Importação de Bibliotecas Básicas

#Análise de Banco de dados
import pandas as pd
import random as random

#k fold sktlearn
from sklearn.model_selection import KFold
```

```python
# Importanto Banco de Dados e retirando as colunas desnecessárias
banco = pd.read_excel('Banco de dados consolidado - Campo.xlsx')
banco = pd.DataFrame(banco)
banco.describe()
```

```python
X, y = banco.drop(['Su-P (kPa)'],axis=1), banco['Su-P (kPa)']

kf = KFold(n_splits=10, shuffle=True, random_state=12345)

for i, (train_index, test_index) in enumerate(kf.split(X, y)):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    X_train.to_excel(f"X_train_fold{i}.xlsx"), X_test.to_excel(f"X_test_fold{i}.xlsx")
    y_train.to_excel(f"Y_train_fold{i}.xlsx"), y_test.to_excel(f"Y_test_fold{i}.xlsx")
```
